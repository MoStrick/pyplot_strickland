1. Given the text below, create a bar plot (or a horizontal bar plot) that shows the frequency distribution of characters in the string. 
A dictionary has been created for you containing the characters and the frequencies of the characters.
Make sure your bar plot has the bars labelled correctly (with the character it represents the frequency of).


```python
from collections import Counter
TEXT = "At three o’clock precisely I was at Baker Street, but Holmes had not yet returned. The landlady informed me that he had left the house shortly after eight o’clock in the morning. I sat down beside the fire, however, with the intention of awaiting him, however long he might be. I was already deeply interested in his inquiry, for, though it was surrounded by none of the grim and strange features which were associated with the two crimes which I have already recorded, still, the nature of the case and the exalted station of his client gave it a character of its own. Indeed, apart from the nature of the investigation which my friend had on hand, there was something in his masterly grasp of a situation, and his keen, incisive reasoning, which made it a pleasure to me to study his system of work, and to follow the quick, subtle methods by which he disentangled the most inextricable mysteries. So accustomed was I to his invariable success that the very possibility of his failing had ceased to enter into my head.".lower()
letter_freq = dict(Counter(TEXT))

# make a bar plot with an appropriate title, correct xtick labels, and labeled axes


# Prepare data for plotting
characters = list(letter_freq.keys())
frequencies = list(letter_freq.values())

# Sort the characters and frequencies based on characters
sorted_characters = sorted(characters)
sorted_frequencies = [letter_freq[char] for char in sorted_characters]


# Determine colors for the bars
colors = []
most_color='#00FF00'  #Green
least_color='#FF00FF'  # Magenta
mid_color='#FFFF00'  # Yellow

sorted_frequencies_with_indices = sorted(enumerate(sorted_frequencies), key=lambda x: x[1])

# Get indices of the top 5 and bottom 5 frequencies
top_5_indices = set(index for index, _ in sorted_frequencies_with_indices[-5:])
bottom_5_indices = set(index for index, _ in sorted_frequencies_with_indices[:5])

# Assign colors based on the frequency ranking
for i in range(len(sorted_frequencies)):
    if i in top_5_indices:
        colors.append(most_color)
    elif i in bottom_5_indices:
        colors.append(least_color)
    else:
        colors.append(mid_color)


# Create a bar plot
plt.figure(figsize=(12, 6))  # Set the figure size
plt.bar(sorted_characters, sorted_frequencies, color=colors)
plt.xlabel('Characters')  # Label for the x-axis
plt.ylabel('Frequency')    # Label for the y-axis
plt.title('Character Frequency Distribution (Highest=Green, Lowest=Magenta)')  # Title of the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(axis='y')  # Add a grid for easier reading of the frequency values
plt.show()
```

2. Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

```python
import matplotlib.pyplot as plt
from random import randint

# Generate a sequence of at least 5 random integers
random_numbers = [randint(1, 100) for _ in range(5)]  # Random integers between 1 and 100

# Prepare x values (indices)
x_values = list(range(len(random_numbers)))

# Create a figure and a set of subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))  # Two vertical subplots

# Scatter plot
ax1.scatter(x_values, random_numbers, color='blue', label='Scatter Plot')
ax1.set_title('Scatter Plot of Random Numbers')
ax1.set_xlabel('Index')
ax1.set_ylabel('Random Number')
ax1.legend()
ax1.grid(True)

# Line plot
ax2.plot(x_values, random_numbers, color='green', marker='o', label='Line Plot')
ax2.set_title('Line Plot of Random Numbers')
ax2.set_xlabel('Index')
ax2.set_ylabel('Random Number')
ax2.legend()
ax2.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

3. Do the same question as above, but change the syle and color of both the scatter and the plot.

```python
import matplotlib.pyplot as plt
from random import randint

#print list of all available fonts
import matplotlib.font_manager as fm
for font in fm.findSystemFonts(fontpaths=None, fontext='ttf'):
    print(fm.FontProperties(fname=font).get_name())


#set font
plt.rcParams.update({'font.family': 'malgun gothic', 'font.size': 12}) 

# Generate a sequence of at least 5 random integers
random_numbers = [randint(1, 100) for _ in range(5)]  # Random integers between 1 and 100

# Prepare x values (indices)
x_values = list(range(len(random_numbers)))

# Calculate the average
average = sum(random_numbers) / len(random_numbers)

# Prepare x values (indices)
x_values = list(range(len(random_numbers)))

# Create a figure and a set of subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))  # Two vertical subplots

# Scatter plot
ax1.scatter(x_values, random_numbers, color='purple', s=100, marker='p')
ax1.set_title(f'Scatter Plot of Random Numbers (Average: {average:.2f})')
ax1.set_xlabel('Index')
ax1.set_ylabel('Random Number')
ax1.legend()
ax1.grid(True)

# Line plot
ax2.plot(x_values, random_numbers, color='lime', linewidth=2)
ax2.scatter(x_values, random_numbers, color='purple', s=150, zorder=5, marker='*')  # Purple star points
ax2.set_title(f'Line Plot of Random Numbers (Average: {average:.2f})')
ax2.set_xlabel('Index')
ax2.set_ylabel('Random Number')
ax2.legend()
ax2.grid(True)
 


#background color
fig.patch.set_facecolor('lightgray')
ax1.set_facecolor('whitesmoke')
ax2.set_facecolor('whitesmoke')

#Change gridlines
ax1.grid(color='lightgray', linestyle='-', linewidth=0.5)  # Light gray, dashed, thin gridlines
ax2.grid(color='lightgray', linestyle='--', linewidth=0.5)  # Light gray, dashed, thin gridlines


# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()

# Print the average to the console
print("Average of the random numbers:", average)
```
